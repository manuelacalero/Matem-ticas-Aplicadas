**4. Crea un programa que pida al usuario Nombre, Peso y Talla y este calcule el
IMC**

```python
# Calcular el IMC

def calcular_imc():
    # Solicitar datos al usuario
    nombre = input("Ingrese su nombre: ")
    peso = float(input("Ingrese su peso en kg: "))
    talla_cm = float(input("Ingrese su altura en cm: "))

    # Convertir talla de cm a metros
    talla = talla_cm / 100

    # Calcular IMC
    imc = peso / (talla ** 2)

    # Mostrar resultado
    print(f"\n{nombre}, tu IMC es: {imc:.2f}")

# Ejecutar la función
calcular_imc()
```

**5. Escriba un programa que pida dos números luego se imprima la suma,
resta, producto y cociente**

```python
# Programa para operaciones

def operaciones_aritmeticas():
    # Solicitar datos al usuario
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))

    # Operaciones
    suma = num1 + num2
    resta = num1 - num2
    producto = num1 * num2
    cociente = num1 / num2 if num2 != 0 else "Indefinido (división por cero)"

    # Mostrar resultados
    print(f"\nSuma: {suma}")
    print(f"Resta: {resta}")
    print(f"Producto: {producto}")
    print(f"Cociente: {cociente}")

# Ejecutar la función
operaciones_aritmeticas()
```

```python
import numpy as np

def calcular_estadisticas(numeros):

    media = np.mean(numeros)
    mediana = np.median(numeros)
    moda = np.bincount(numeros).argmax()

    return media, mediana, moda
numeros = [1, 2, 2, 3, 4, 5, 5, 5]
calcular_estadisticas(numeros)
```
